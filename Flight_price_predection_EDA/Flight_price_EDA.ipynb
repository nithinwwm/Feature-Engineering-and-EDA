```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
#get basic info
df.info()
```

```python
df.columns
```

```python
#cleaning every column for model training
#1st column -- Date_of_Journey
#We will be separating the date to -- date , month and year
```

```python
#Feature Eng
df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
df.info()
```

```python
#but the 3 new columns are still in object type, so we have to conver it into int type
df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey',inplace=True, axis=1)
df.head()
```

```python
##Now clean Arrival_Time

#df['Arrival_hour'] = df['Arrival_Time'].str.split().str[0].str.split(':').str[0].astype(int)
#df['Arrival_minute'] = df['Arrival_Time'].str.split().str[0].str.split(':').str[1].astype(int)

##This is simple but lenghty
##To get the same results, lets use labmda
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split()[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

```python
#Now we will take care of Departure time
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_minute'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
df.info()
```

```python
df.head()
```

```python
#Now we will clean Total_Stops column
df['Total_Stops'].unique()

#There is nan value, we will replace it actual value and we have map/encode the other values to numeric values 
df[df['Total_Stops'].isnull()]

#We will replace the nan value with mode value and others to numeric values
df['Total_Stops'].mode()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df['Total_Stops'].isnull().sum()
df.head()
```

```python
#Now we can drop Route since we have source and destination already
df.drop('Route',axis=1,inplace=True)
df.head()
```

```python
#Now we clean the 'Duration' column
df['Duration_hours'] = df['Duration'].str.split('h').str[0]
df['Duration_hours'].unique()
##we need to take care of '5m' found in Duration_hours
df[df['Duration_hours']=='5m']
#First we will create duration_minutes column
df['Duration_minutes'] = df['Duration'].str.split('h').str[1].str.split('m').str[0]
df.info()
```

```python
#We need to replace duration_hours to 0 and Duration_minutes to 5
df.loc[df['Duration']=='5m', 'Duration_minutes'] = 5
df.loc[df['Duration']=='5m', 'Duration_hours'] = 0
df[df['Duration']=='5m']
```

```python
print(df['Duration_hours'].unique())
print(df['Duration_minutes'].unique())
df['Duration_hours'] = df['Duration_hours'].astype(int)
```

```python
df.loc[df['Duration_minutes'] == '', 'Duration_minutes'] = 0
df['Duration_minutes'] = df['Duration_minutes'].astype(int)
df['Duration_minutes'].unique()
df.drop('Duration',axis=1,inplace=True)
df.info()
```

```python
#We have taken care of all columns except categorical columns
df['Airline'].unique()
#We use Onehotencoding for Airline, Source and Destination columns
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray()
#Now this is in array format, we need to convert it into dataframe and concatenate with original dataframe
```

```python
encoded_df = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray(),columns=encoder.get_feature_names_out())
encoded_df.shape
```

```python
final_df = pd.concat([df,encoded_df],axis=1)
final_df.drop(['Airline','Source','Destination','Additional_Info'],axis=1,inplace=True)
final_df.info()
final_df.head()
```
